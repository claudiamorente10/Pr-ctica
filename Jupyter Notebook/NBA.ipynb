```python
import pandas as pd

NBA = pd.read_csv('C:/master/Pràctica II Visualització/NBA.csv')
```

```python
NBA.head(5)
```

```python
NBA = NBA.drop_duplicates()

NBA.head(5)
```

```python
NBA['NICKNAME'].unique()
```

```python
nba = NBA.groupby(['NICKNAME', 'YEAR', 'CITY']).agg({'PTS_home': ['sum', 'mean'], 'PTS_away': ['sum', 'mean'], 'AST_home': ['sum', 'mean'], 'AST_away': ['sum', 'mean'], 'REB_home': ['sum', 'mean'], 'REB_away': ['sum', 'mean'], 'FG_PCT_home': 'mean', 'FG_PCT_away': 'mean', 'FT_PCT_home': 'mean', 'FT_PCT_away': 'mean', 'FG3_PCT_home': 'mean', 'FG3_PCT_away': 'mean'})

nba.columns = ['Total PTS_home', 'Mean PTS_home', 'Total PTS_away', 'Mean PTS_away', 'Total AST_home', 'Mean AST_home', 'Total AST_away', 'Mean AST_away', 'Total REB_home', 'Mean REB_home', 'Total REB_away', 'Mean REB_away', 'FG_PCT_home', 'FG_PCT_away', 'FT_PCT_home', 'FT_PCT_away', 'FG3_PCT_home', 'FG3_PCT_away']
nba = nba.reset_index()

nba.head(6)
```

```python
points = nba['Total PTS_home'] + nba['Total PTS_away']
mean_points = (nba['Mean PTS_home'] + nba['Mean PTS_away'])/2
assists = nba['Total AST_home'] + nba['Total AST_away']
mean_assists = (nba['Mean AST_home'] + nba['Mean AST_away'])/2
rebounds = nba['Total REB_home'] + nba['Total REB_away']
mean_rebounds = (nba['Mean REB_home'] + nba['Mean REB_away'])/2
FG_PCT = (nba['FG_PCT_home'] + nba['FG_PCT_away'])/2
FT_PCT = (nba['FT_PCT_home'] + nba['FT_PCT_away'])/2
FG3_PCT = (nba['FG3_PCT_home'] + nba['FG3_PCT_away'])/2

nba['Total PTS'] = points
nba['Mean PTS'] = mean_points
nba['Total AST'] = assists
nba['Mean AST'] = mean_assists
nba['Total REB'] = rebounds
nba['Mean REB'] = mean_rebounds
nba['Field Goal %'] = FG_PCT
nba['Free Throw %'] = FT_PCT
nba['3-point Field Goal %'] = FG3_PCT

nba.head(6)
```

```python
nba.drop('Total PTS_home', inplace=True, axis=1)
nba.drop('Mean PTS_home', inplace=True, axis=1)
nba.drop('Total PTS_away', inplace=True, axis=1)
nba.drop('Mean PTS_away', inplace=True, axis=1)
nba.drop('Total AST_home', inplace=True, axis=1)
nba.drop('Mean AST_home', inplace=True, axis=1)
nba.drop('Total AST_away', inplace=True, axis=1)
nba.drop('Mean AST_away', inplace=True, axis=1)
nba.drop('Total REB_home', inplace=True, axis=1)
nba.drop('Mean REB_home', inplace=True, axis=1)
nba.drop('Total REB_away', inplace=True, axis=1)
nba.drop('Mean REB_away', inplace=True, axis=1)
nba.drop('FG_PCT_home', inplace=True, axis=1)
nba.drop('FG_PCT_away', inplace=True, axis=1)
nba.drop('FT_PCT_home', inplace=True, axis=1)
nba.drop('FT_PCT_away', inplace=True, axis=1)
nba.drop('FG3_PCT_home', inplace=True, axis=1)
nba.drop('FG3_PCT_away', inplace=True, axis=1)
```
